### Python

```python
# Load libraries
import pandas as pd
import xlrd
import matplotlib.pyplot as plt
```

```python
# Read data file
try:
    population_df = pd.read_excel("data/world-population.xlsm")
except Exception:
    print("There was an error reading the World Population file")
    
population_df.head()
```

#### Charts

**1. Line Chart**

```python
plt.rcdefaults()
fig, ax = plt.subplots(figsize=(10, 6))

# plot bar chart
ax.plot(population_df['Year'], (population_df['Population'] / 1000000000))

# set labels
ax.set_ylabel('Population (in Billions)', fontsize = 14)
ax.ticklabel_format(useOffset = False, style = 'plain')
ax.set_title('Line Chart - Python \nPopulation per Year: 1960 - 2010', pad = 20)
ax.yaxis.labelpad = 20.0

plt.show()

# Save figure
ax.get_figure().savefig('images/linechart-python.png',
           bbox_inches = 'tight',
           transparent = True)
```

**2. Step Chart**

```python
plt.rcdefaults()
fig, ax = plt.subplots(figsize=(10, 6))

# plot bar chart
ax.step(population_df['Year'], (population_df['Population'] / 1000000000))

# set labels
ax.set_ylabel('Population (in Billions)', fontsize = 14)
ax.ticklabel_format(useOffset = False, style = 'plain')
ax.set_title('Step Chart - Python \nPopulation per Year: 1960 - 2010', pad = 20)
ax.yaxis.labelpad = 20.0

plt.show()

# Save figure
ax.get_figure().savefig('images/stepchart-python.png',
           bbox_inches = 'tight',
           transparent = True)
```
