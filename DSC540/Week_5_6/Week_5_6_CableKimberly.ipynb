#### Manipulate Strings

##### Just keep search number in Link Resource since the URL is the same in all

```python
met_cleaned['Link Resource'][100:120]
```

```python
# Split out URL and only keep the last node 
met_cleaned['Link Resource Num'] = met_cleaned['Link Resource'].apply(lambda x: x.split('/')[-1])

# Drop original column
met_cleaned.drop(['Link Resource'], axis = 1, inplace = True)

# Check
met_cleaned.head()
```

##### Object Wikidata URL: just keep object wiki since the URL is the same

```python
met_cleaned.loc[met_cleaned['Object Wikidata URL'].notnull()].head()
```

```python
# Fil in empty rows with Unknown
met_cleaned['Object Wikidata URL'] = met_cleaned['Object Wikidata URL'].fillna('Unknown')

# Split out URL and only keep the last node 
met_cleaned['Object Wiki'] = met_cleaned['Object Wikidata URL'].apply(lambda x: x.split('/')[-1])

# Drop original column
met_cleaned.drop(['Object Wikidata URL'], axis = 1, inplace = True)

# Check
met_cleaned.loc[met_cleaned['Object Wiki'] != 'Unknown'].head()
```

```python
# Split dataframe into different Department Dataframes (for merging)
unique_departments = met_cleaned['Department'].unique().tolist()
unique_departments
```
