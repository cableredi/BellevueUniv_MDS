```python
met_cleaned['Object Start Date'] = pd.to_datetime(met_cleaned['Object Begin Date'])
met_cleaned['Object Complete Date'] = pd.to_datetime(met_cleaned['Object End Date'])
met_cleaned.info()
```

#### Generate Date Range

```python
def create_range(row):
    begin = row.strftime('%Y-%m-%d')
    
    range = pd.date_range(begin, periods = 4)   
    return range

met_cleaned['range'] = met_cleaned['Object Start Date'].apply(lambda row: create_range(row))
met_cleaned.head()
```
